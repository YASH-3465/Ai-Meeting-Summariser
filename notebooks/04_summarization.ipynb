```python
from transformers import pipeline
```

```python
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn",
    device=-1   # CPU
)
```

```python
def chunk_text(text, max_words=400):
    words = text.split()
    chunks = []

    for i in range(0, len(words), max_words):
        chunk = " ".join(words[i:i + max_words])
        chunks.append(chunk)

    return chunks
```

```python
cleaned = """
good morning everyone. thank you for joining the weekly project review meeting.
today we will discuss the current project status timeline risks and next action items.
the main focus of this meeting is to review progress made during the last sprint.

the development team has completed the user authentication module and the dashboard ui.
api integration for user profiles is currently in progress and expected to be completed by friday.
testing for completed modules has started and initial feedback has been positive.

however there are some challenges related to database performance and query optimization.
the backend team is working on indexing and caching strategies to improve response time.
this issue may slightly affect the original timeline if not resolved by the end of the week.

regarding the project timeline the development phase is planned to end in two weeks.
testing and bug fixing will take an additional one week.
the deployment to staging environment is scheduled for the following monday.
production release is tentatively planned for the end of this month.

the total project budget is approximately 50000.
additional infrastructure costs may arise if performance optimization requires cloud scaling.
budget approval for any extra expenses must be communicated to management in advance.

john is responsible for backend development and performance optimization.
priya is managing testing and quality assurance.
arjun is handling frontend development and ui improvements.
neha will coordinate deployment and monitoring activities.

there was also a discussion about improving documentation and onboarding guides.
the team agreed that better documentation will help new developers understand the system faster.
documentation updates will be handled alongside development work.

the next meeting is scheduled for monday at 11 am.
by then all teams are expected to share updated progress reports.
any blockers or delays should be communicated before the next meeting.

thank you everyone for your contributions and continued effort.
let us stay aligned on deadlines and maintain communication across teams.
"""
```

```python
chunks = chunk_text(cleaned,max_words=150)

print("Total chunks:", len(chunks))
```

```python
chunk_summaries = []

for i, chunk in enumerate(chunks):
    summary = summarizer(
        chunk,
        max_length=120,
        min_length=40,
        do_sample=False
    )
    chunk_summaries.append(summary[0]["summary_text"])
    print(f"Chunk {i+1} summarized")
```

```python
final_summary = " ".join(chunk_summaries)

print(final_summary)
```
