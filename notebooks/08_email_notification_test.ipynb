```python
sample_actions = [
    {
        "action": "John will prepare the final project report",
        "deadline": "by Friday"
    },
    {
        "action": "The team must submit the demo video",
        "deadline": "12/10/2025"
    }
]
```

```python
import smtplib
from email.mime.text import MIMEText

def send_test_email(actions, to_email):
    body = ""
    for a in actions:
        body += f"- {a['action']}\nDeadline: {a['deadline']}\n\n"

    msg = MIMEText(body)
    msg["Subject"] = "Meeting Action Items"
    msg["From"] = "yashwanthgoud2007@gmail.com"
    msg["To"] = to_email

    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()
    server.login("yashwanthgoud2007@gmail.com", "umakdojbdkxwwxag")
    server.send_message(msg)
    server.quit()
```

```python
send_test_email(sample_actions, "komaramadhavarao@gmail.com")
```
