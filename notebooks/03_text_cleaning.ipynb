```python
import re

def clean_text(text):
    text = text.lower()

    text = re.sub(r'\[.*?\]|\(.*?\)', ' ', text)

    text = re.sub(r'\b(uh+|um+|umm+|ah+|like|you know|yeah+)\b', ' ', text)

    text = re.sub(r'(.)\1{2,}', r'\1', text)

    text = re.sub(r'\S+@\S+', ' ', text)

    text = re.sub(r'\+?\d[\d\s\-]{8,}', ' ', text)

    text = re.sub(r'[^a-z0-9\s\.\,]', ' ', text)

    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
raw_text = """
Uh ummm good morning everyone!!!
(laughs) Today’s meeting is on 12/12/2024 @ 10:30 AM.
Contact me at yash123@gmail.com or +91-98765-43210
yeahhhh let's start!!!
"""

print("RAW TEXT:\n", raw_text)
print("\nCLEANED TEXT:\n", clean_text(raw_text))
```
